```python
df = pd.read_csv('/StudentsPerformance.csv')
average_scores = df.groupby('gender').mean()
average_scores.reset_index(inplace=True)
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.3
index = range(len(average_scores))
bar1 = ax.bar(index, average_scores['math score'], bar_width, label='Math')
bar2 = ax.bar([i + bar_width for i in index], average_scores['reading score'], bar_width, label='Reading')
bar3 = ax.bar([i + 2 * bar_width for i in index], average_scores['writing score'], bar_width, label='Writing')
ax.set_xlabel('gender')
ax.set_ylabel('Average Score')
ax.set_title('Average Scores by gender and Test')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(average_scores['gender'])
ax.legend()
plt.show()
```

```python
df = pd.read_csv('/StudentsPerformance.csv')
plt.figure(figsize=(10, 6))
sns.violinplot(x='parental level of education', y='math score', data=df, inner='quart', cut=0)
plt.xlabel('Parental Education')
plt.ylabel('Math Score')
plt.title('Math Score Distributions by Parental Education')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='parental level of education', y='reading score', data=df, inner='quart', cut=0)
plt.xlabel('Parental Education')
plt.ylabel('Reading Score')
plt.title('Reading Score Distributions by Parental Education')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='parental level of education', y='writing score', data=df, inner='quart', cut=0)
plt.xlabel('Parental Education')
plt.ylabel('Writing Score')
plt.title('Writing Score Distributions by Parental Education')
plt.xticks(rotation=45)
plt.show()
```
